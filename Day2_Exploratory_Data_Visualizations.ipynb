```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("===== Exploratory Data Visualization =====")

# Sample Dataset
data = {
    "Age": [22,25,28,30,35,40,45,50,29,32],
    "Salary": [25000,30000,35000,40000,50000,60000,65000,70000,42000,48000],
    "Experience": [1,2,3,5,7,10,12,15,4,6]
}

df = pd.DataFrame(data)

print(df)

# -----------------------------
# Scatter Plot (Visual Correlation)
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.savefig("scatter_plot.png")
plt.show()

# -----------------------------
# Correlation Heatmap
# -----------------------------
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.savefig("heatmap.png")
plt.show()

# -----------------------------
# Distribution Curve (Histogram)
# -----------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["Salary"], kde=True)
plt.title("Salary Distribution")
plt.savefig("distribution_curve.png")
plt.show()

# -----------------------------
# Bar Chart
# -----------------------------
plt.figure(figsize=(6,4))
plt.bar(df["Age"], df["Salary"])
plt.title("Age vs Salary Bar Chart")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.savefig("bar_chart.png")
plt.show()

print("\nCharts saved successfully!")
```
