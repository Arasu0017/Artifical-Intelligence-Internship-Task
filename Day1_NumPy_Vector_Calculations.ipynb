```python
import numpy as np

print("===== NumPy Vector Calculations =====")

# Create vectors
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("\nVector A:", A)
print("Vector B:", B)

# Addition
print("\nVector Addition:")
print(A + B)

# Subtraction
print("\nVector Subtraction:")
print(A - B)

# Dot Product
print("\nDot Product:")
print(np.dot(A, B))

# Matrix Creation
M1 = np.array([[1, 2],
               [3, 4]])

M2 = np.array([[5, 6],
               [7, 8]])

print("\nMatrix M1:")
print(M1)

print("\nMatrix M2:")
print(M2)

# Matrix Multiplication
print("\nMatrix Multiplication:")
print(np.matmul(M1, M2))

# Matrix Transpose
print("\nTranspose of M1:")
print(M1.T)

# Identity Matrix
print("\nIdentity Matrix:")
print(np.eye(3))

# Inverse Matrix
print("\nInverse of M1:")
print(np.linalg.inv(M1))

# Determinant
print("\nDeterminant of M1:")
print(np.linalg.det(M1))

# Eigenvalues and Eigenvectors
values, vectors = np.linalg.eig(M1)

print("\nEigenvalues:")
print(values)

print("\nEigenvectors:")
print(vectors)
```
